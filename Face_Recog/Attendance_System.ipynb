```python
import face_recognition

import numpy as np
import pandas as pd
import csv
import cv2
import glob

import os
from datetime import datetime
```

```python
# EYES AND FACE


face_cap = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eyes_cap = cv2.CascadeClassifier('haarcascade_eye.xml')
video_cap = cv2.VideoCapture(0)

while True :
    ret , video_data = video_cap.read()
    col = cv2.cvtColor(video_data,cv2.COLOR_BGR2GRAY)
    
    faces = face_cap.detectMultiScale(
            col,
            scaleFactor = 1.1,
                minNeighbors = 5,
                minSize = (30,30),
                flags = cv2.CASCADE_SCALE_IMAGE
            )

    eyes = eyes_cap.detectMultiScale(
        col,
        scaleFactor = 1.1,
            minNeighbors = 5,
            minSize = (30,30),
            flags = cv2.CASCADE_SCALE_IMAGE
        )
        
    for (x,y,w,h) in faces:
         cv2.rectangle(video_data,(x,y),(x+w,y+h),(255,0,0),2)    # (255,0,0) - BGR color- change as you will. width = 2
            
    for (x,y,w,h) in eyes:
        cv2.rectangle(video_data,(x,y),(x+w,y+h),(0,225,0),2)    # (0,225,0) - BGR color- change as you will. width = 2
    
    cv2.imshow("Video_Live",video_data)
    if cv2.waitKey(5)==ord('q'):
        break
video_cap.release()   
cv2.destroyAllWindows()    # to close all windows
```
